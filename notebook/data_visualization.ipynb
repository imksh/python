**Matplotlib**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr1=np.linspace(0,5,11)
arr2=arr1**2
print(arr1)
print(arr2)
```

basics

```python
plt.plot(arr1,arr2)
plt.title("graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

subplots

divides the canvas into rows and cols 
3 arguments row,col,index of current

```python
plt.subplot(2,2,1)
plt.plot(arr1,arr1)
plt.subplot(2,2,4)
plt.plot(arr1,arr2)
plt.subplot(2,2,3)
plt.plot(arr2,arr1)
plt.subplot(2,2,2)
plt.plot(arr1,arr2**2)
```

Object oriented way

```python
"""
matplolib allows to set figure size passed as argument as tupple(width,height)  and dpi(depth per inch)default = 100
fig=plt.figure(figsize=(12,8),dpi=500)
"""

fig=plt.figure()
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(arr1,arr2)
axis2 = fig.add_axes([0.3,0.3,0.2,0.2])
axis2.plot(arr1,arr2)



#add_axes creates new axis 4 arguments 1. left 2. bottom 3. width 4. height  between 0 and 1
```

scatter plot

```python
plt.scatter(arr1,arr2)
```

histogram

```python
plt.hist(arr1)

#better in seabon
```

box plot

```python
plt.boxplot(arr1)
```

saving plot

```python
plt.plot(arr1,arr2)
plt.title("graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.savefig("fig1")
```

**Distributation Plot**

used for quantative data (like numbers ,age)
example:-
1. distplot/histplot
2. jointplot
3. pairplot
4. rugplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")
df
```

```python
df["size"].unique()
```

```python
#distplot

sns.displot(df["total_bill"])

#bins are created on x axis and data falling in that bin accumalate there and count is shown on y axis
#bins by default = 20 bins
```

```python
sns.displot(df["total_bill"], bins=2)
```

```python
sns.histplot(df["total_bill"], bins=2)
```

```python
sns.displot(df["total_bill"],kde=True)

#kde=kernal density estimeter
```

```python
#2 plot in one canvas

plt.subplot(1,2,1)
sns.histplot(df["total_bill"], kde=True)
plt.subplot(1,2,2)
sns.histplot(df["tip"], kde=True)
```

jointplot

to show to data

```python
sns.jointplot(x="total_bill",y="tip",data=df)

#kind = scatter (default)
```

```python
sns.jointplot(x="total_bill",y="tip",data=df,kind="hex")
```

```python
sns.jointplot(x="total_bill",y="tip",data=df,kind="reg")
```

```python
sns.jointplot(x="total_bill",y="tip",data=df,kind="kde")
```

pairplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="sex")
```
